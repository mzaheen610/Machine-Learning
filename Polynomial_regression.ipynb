```python
df = pd.read_csv('Salary_Data.csv')
df.shape
df.info()
df.describe()
```

```python
df.head()
```

```python
#fitting the model
X, Y = df["YearsExperience"] , df['Salary']
# print(X,Y)
linear_model = np.poly1d(np.polyfit(X, Y, 1))
print(linear_model)
```

```python
#plotting the linear model
plt.scatter(X, Y)
my_line = np.linspace(1, 15, 100)
plt.plot(my_line, linear_model(my_line))
```

```python
#plotting the 2d model
poly_model = np.poly1d(np.polyfit(X, Y, 3))
print(poly_model)
plt.scatter(X, Y)
my_line = np.linspace(1, 15, 100)
plt.plot(my_line, poly_model(my_line))
```

```python
#calculating the metrics for the model
from sklearn.metrics import r2_score, mean_squared_error
score_poly = r2_score(Y, poly_model(X))
score_linear = r2_score(Y, linear_model(X))
print(score_linear)
print(score_poly)
```
